```python
#importing decision tree, logistic regression and KNN

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier
```

```python
# importing required packages and libraries 
import pandas as pd
import numpy as np
```

# Part A

```python
df_heart = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")
```

```python
#priting the df for my knowledge
df_heart
```

# Part C

```python
#creating python list of feature names (non catagorical)
feature_cols = ['Age' , 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak']
#use above list to select features from original dataframe
X = df_heart[feature_cols]
#printing first 5 rows
X.head()
```

```python
#selecting last column from dataframe
y = df_heart['AHD']
#checking label vector by printing every 20 values
y[::20]
```


# Part D

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=6)
```

# Part E

```python
#KNN k=3
k=3
heart = KNeighborsClassifier(n_neighbors=k)
```

```python
#using the fit method to train the data and labels for KNN
heart.fit(X,y)
```

```python
#training on training set for KNN
heart.fit(X_train, y_train)
```

```python
#testing on testing set for KNN
y_predict = heart.predict(X_test)
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)

print(accuracy)
```

```python
#DecisionTree and Logistic Regression
LR = LogisticRegression()
DT = DecisionTreeClassifier(random_state=5)
```

```python
#training LR and DT models
LR.fit(X,y)
DT.fit(X,y)
```

```python
#training the models
LR.fit(X_train,y_train)
DT.fit(X_train, y_train)
```

```python
#Testing data on testing set
y_predict1 = LR.predict(X_test)
y_predict2 = DT.predict(X_test)
```

```python
#logistic regression with random_state=6
accuracy_LR = accuracy_score(y_test, y_predict1)
print(accuracy_LR)

accuracy_DT = accuracy_score(y_test, y_predict2)
print(accuracy_DT)
```

# Which one is the best?
KNN, k = 3 has a 64.5% accuracy
LR, has a 67% accuracy
DT, random_state=5 has a 61.8% accuracy.
From the numbers that I have recieved it seems that logistic regression is the most accurate model with an accuracy of 67%

```python
#creating dummies for gender
gender = pd.get_dummies(df_heart.Gender)
gender
```

```python
y_predictf = fin.predict(X_testf)
```

```python
acc = accuracy_score(y_testf,y_predictf)
acc
```

```python
#creating objects for DT and LR
my_logreg = LogisticRegression()
my_decisiontree = DecisionTreeClassifier(random_state=5)
```

```python
my_decisiontree.fit(X1,y1)
```

```python
# Testing on the testing set:

y_predict_lr = my_logreg.predict(X_testf)

y_predict_dt = my_decisiontree.predict(X_testf)
```

```python
score_lr = accuracy_score(y_test, y_predict_lr)
score_dt = accuracy_score(y_test, y_predict_dt)

print(score_lr)
print(score_dt)
```

The new data inputs seemed to increase the accuracy a little. KNN accuracy stayed the same, LR accuracy increased by 10% and DT accuracy increased by 12%.   

# Part H

```python
# importing cross validation method:
from sklearn.model_selection import cross_val_score
```

```python
#using same features as part F
features = ['Age' , 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak','f','m','asymptomatic','nonanginal','nontypical','typical','fixed','normal','reversable']

X = final[features]

y = final['AHD']

X
```

```python
#setting k = 3 for KNN
k = 3
my_KNN = KNeighborsClassifier(n_neighbors=k)
```

```python
#creating a list for cross val scores
accuracy_KNN = cross_val_score(my_KNN, X, y, cv=10, scoring='accuracy')

print(accuracy_KNN)
```

```python
#Printing accuracy for CV on KNN where K = 3
accuracy_K = accuracy_KNN.mean()

print(accuracy_K)
```

```python
#Applying CV with DT and random state = 5
my_DT = DecisionTreeClassifier(random_state=5)
#creating object of DT to hold CV scores
accuracy_DT = cross_val_score(my_DT, X, y, cv=10, scoring='accuracy')
#printing CV scores
print(accuracy_DT)
```

```python
#Creating varibale to hold mean of all DT accuracies 
accuracyD = accuracy_DT.mean()

print(accuracyD)
```

```python
#starting to apply CV with logistic regression
my_LR = LogisticRegression(max_iter = 10000)
#creating list to store cv values from LR
accuracy_LR = cross_val_score(my_LR, X, y, cv = 10, scoring = 'accuracy')
#printing LR CV score list
print(accuracy_LR)
```

```python
#creating object to store mean of LR score values
accuracyL = accuracy_LR.mean()
#printing out accuracy of LR and CV 
print(accuracyL)
```
